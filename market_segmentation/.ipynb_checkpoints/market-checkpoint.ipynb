```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
data = pd.read_csv('social_marketing.csv')
#check distrubution of the data
data.isnull().sum()
```

```python
data.info()
```

```python
#print first 5 lines of the data
data[:5]
```

```python
# Drop the user identifier column if it's present
if 'user_id' in data.columns:
    data = data.drop(columns=['user_id'])

# Calculate the correlation matrix
correlation_matrix = data.corr(numeric_only=True)  # Set numeric_only=True

# Create a heatmap of the correlation matrix using seaborn
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Matrix of User Interests")
plt.show()

```

```python
#drop 'chatter, 'spam', 'adult' columns
data = data.drop(columns = ['chatter', 'spam', 'adult','Unnamed: 0'])
```

```python
# Elbow method to find the optimal number of clusters
inertia = []
for k in range(1, 11):  # trying different numbers of clusters
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0).fit(data)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()


```

```python
# Standardize the data
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data)

# Perform k-means clustering
kmeans = KMeans(n_clusters=5, n_init=10, random_state=0)
kmeans.fit(standardized_data)
y_kmeans = kmeans.predict(standardized_data)


# Plot the data points and cluster centers
plt.scatter(standardized_data[:, 0], standardized_data[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.title('cluster')
plt.show()


```

```python

# Add cluster labels to the original DataFrame
data['Cluster'] = y_kmeans

# Calculate mean values for each feature within each cluster
cluster_summary = data.groupby('Cluster').mean()

# Print cluster information
print(cluster_summary)
```
